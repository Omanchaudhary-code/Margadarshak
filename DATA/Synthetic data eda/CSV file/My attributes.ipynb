```python
categorical_cols = [
    "Average sleep per night",
    "Do you feel supported by friends or family in your academics?",
    "Estimated monthly household income (NPR)",
    "Are you the first in your family to attend university?"
]
```

```python
for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.xticks(rotation=45)
    plt.title(f"Distribution of {col}")
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df["How is your friend circle academically?"], bins=5, kde=True)
plt.title("Academic Strength of Friend Circle")
plt.tight_layout()
plt.show()
```
